## Findings
### Time Trends of Airbnb

##### Separating data by pandemic years
The Seattle Airbnb data runs from 2019 to March 2023. To analyze the COVID factor in Airbnb, I will separate the data by the year 2019 for pre-covid and 2020 & beyond for COVID.

### IMPORTANT NOTE - Cleaning data for the years
For answering the questions about changes in Airbnb after COVID began, the only source of time is the ``last_review`` variable. This variable indicates the date when the person reviews their stay. I am assuming that people review the Airbnb relatively close in time to when they stayed in the Airbnb. Furthermore, there is a significant portion of missing data for this variable so I removed 1087 (out of 6376) rows where ``last_review`` is empty. 

```python
# Take out empty last_review rows
clean_data = []

for row in bnb_data:
    if row["last_review"] != "":
        clean_data.append(row)

print(len(clean_data))
```

```python
# Separate data each year by storing rows iin a list
precovid_data = []
data2020 = []
data2021 = []
data2022 = []
data2023 = []

for row in clean_data:
    if int(row["last_review"][:4]) <= 2019:
        precovid_data.append(row)
    elif int(row["last_review"][:4]) == 2020:
        data2020.append(row)
    elif int(row["last_review"][:4]) == 2021:
        data2021.append(row)
    elif int(row["last_review"][:4]) == 2022:
        data2022.append(row)
    elif int(row["last_review"][:4]) == 2023:
        data2023.append(row)
        
print(len(precovid_data))
print(len(data2020))
print(len(data2021))
print(len(data2022))
print(len(data2023))
```

I made a function called ``make_list`` to create a list of the values based on the given attribute you pass as a parameter in the function
- ``list_name`` is a list you pass in for the function to store the values (usually just an empty list)
- ``last_review`` is the name of the column header (variable) that you are interested in
- ``data`` is the dataset to use

```python
# Make a list of chosen attributes
def make_list(list_name, attribute, data):
    list_name = []
    for row in data:
        list_name.append(int(row.get(attribute)))
    return list_name
```

```python
# Package to calculate statistics
import statistics
```

```python
# Use the function I created above to make lists for all the years in the datset
years = ["2019", "2020", "2021", "2022", "2023"]
price_year_avg = []

year2019 = make_list([], "price", precovid_data)
price_year_avg.append(statistics.mean(year2019))

year2020 = make_list([], "price", data2020)
price_year_avg.append(statistics.mean(year2020))

year2021 = make_list([], "price", data2021)
price_year_avg.append(statistics.mean(year2021))

year2022 = make_list([], "price", data2022)
price_year_avg.append(statistics.mean(year2022))

year2023 = make_list([], "price", data2023)
price_year_avg.append(statistics.mean(year2023))

print(price_year_avg)
```

```python
# Create line graph for price over years
plt.plot(years, price_year_avg, color='red', marker='o')
plt.title('Price vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.grid(True)

plt.show()
```

This result is surprising to me. It makes sense that the prices decrease when COVID hit in 2020. But I expected to see the prices beginning to increase from 2022 and beyond because I assume people will begin traveling again. Also with inflation, I expected everything to get more expensive so I thought that the listing prices would increase over time.

For the next few questions I made a function to count the different attributes of the dataset to reduce redundancy.
Steps I took to make this function:
- Pass in 3 parameters
    - 1st param is the empty dictionary
    - 2nd param is the characteristic of the rows we are counting
    - dataset is the name of the dataset
- Loop through the dataset
    - if the key is already in the dictionary, we increase the value by one
    - otherwise if the dictionary does not contain the key, we create a new key/pair with a value of 1

```python
# Make a function to count amount of Airbnbs in neighborhoods
def count_attr(bnb_dict, attribute, dataset):
    for row in dataset:
        if row[attribute] in bnb_dict:
            bnb_dict[row[attribute]] += 1
        else:
            bnb_dict[row[attribute]] = 1
    return(bnb_dict)
```

```python
# See the number of Airbnbs by neighborhood
neighborhood_dict = count_attr({}, "neighbourhood_group", bnb_data)
print(neighborhood_dict)
```

```python
# Plot a bar chart
plt.bar(range(len(neighborhood_dict)), list(neighborhood_dict.values()), align='center')
plt.suptitle('Airbnb in Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Airbnbs')
plt.xticks(range(len(neighborhood_dict)), list(neighborhood_dict.keys()), rotation = 50)
plt.gcf().set_size_inches(10, 5)

plt.show()
```

I couldn't get the x labels to align exactly with the appropriate bars. Another thing is that the highest bar is "Other neighborhoods" but I will disregard it because is not in any of the other neighborhood categories. So the neighborhood with the most Airbnb is Downtown followed by Central Area and Capitol Hill. These results make sense because those areas are where tourists frequent the most. Also, the neighborhoods with fewer Airbnbs are Interbay, followed by Seward Park and Magnolia.

In the next section, I made a function to calculate the average:
- Pass in 3 parameters
    - 1st param is the name of the neighborhood I want to find the avergae of
    - 2nd param is the characteristic (column) of the Airbnb
    - data is the name of the dataset
- Loop through the dataset
    - if the name of the neigborhood matches the name in the dataset, add it to the total amount
- Divide the cumulative amount by the total number of Aribnbs in the neighborhood

```python
# Make a function to calculate the average listing price in neighborhoods
def calc_avg(neighborhood_name, attribute, data):
    total_price = 0
    total_bnbs = neighborhood_dict.get(neighborhood_name)  # neighborhood_dict has the amount of Airbnbs by neighborhood

    for row in data:
        if row["neighbourhood_group"] == neighborhood_name:
            total_price += int(row[attribute])

    avg = total_price / total_bnbs
    return(avg)
```

```python
# Put neighborhoods and price per day in a dictionary
neighborhood_price = {}

for neighborhood in neighborhood_dict:
    neighborhood_price[neighborhood] = round(calc_avg(neighborhood, "price", bnb_data), 2)
    
print(neighborhood_price)
```

```python
# Plot a bar chart of neighborhood and Airbnb average prices
plt.bar(range(len(neighborhood_price)), list(neighborhood_price.values()), align='center')
plt.suptitle('Average Airbnb Prices by Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Price (USD)')
plt.xticks(range(len(neighborhood_price)), list(neighborhood_price.keys()), rotation = 50)
plt.gcf().set_size_inches(10, 5)

plt.show()
```

This result kind of surprised me. I assumed that the neighborhoods that I mentioned above (with the most amounts) would be more expensive because of the tourist spots and the competition of Airbnb. But perhaps I made a mistake somewhere in my code when I calculated the averages.

### For the next part, I am interested to see the popularity of the Airbnbs

To examine this popularity aspect I will be assuming that the number_of_reviews means it is more popular. I would think that the more reviews an Airbnb received, means people enjoyed their stay enough to put a review and more reviews would make people more likely to book an Airbnb. Even if the review is not positive, more reviews mean at least a lot of people have stayed in that Airbnb.

```python
# Put neighborhoods and number of reviews in a dictionary
reviews_dict = {}

# Add up the toal number of reviews by neighborhoods
for neighborhood in bnb_data:
    if neighborhood["neighbourhood_group"] in reviews_dict:
        reviews_dict[neighborhood["neighbourhood_group"]] += int(neighborhood["number_of_reviews"])
    else:
        reviews_dict[neighborhood["neighbourhood_group"]] = int(neighborhood["number_of_reviews"])
        
print(reviews_dict)
```

```python
# Plot a bar chart of neighborhood and Airbnb average reviews
plt.bar(range(len(reviews_dict)), list(reviews_dict.values()), align='center')
plt.suptitle('Total Airbnb Reviews by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Reviews')
plt.xticks(range(len(reviews_dict)), list(reviews_dict.keys()), rotation = 50)
plt.gcf().set_size_inches(10, 5)

plt.show()
```

This result is similar to the graph with the number of Airbnbs in Seattle nieghborhoods. It makes sense that Airbnbs within the downtown area will have more reviews written by people becuase there are more Airbnbs in those neighborhoods.
